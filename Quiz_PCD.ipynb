# **NOMOR 1**
**baca gambar menggunakan opencv dan PIL**

```python
# Nomor 1
from torchvision import transforms
import cv2
from google.colab.patches import cv2_imshow
import PIL.Image as Image
from skimage.util import random_noise

# Membaca gambar menggunakan OpenCV
img = cv2.imread('/content/quiz.jpeg')
img = cv2.resize(img,(400, 400))
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Mengonversi gambar OpenCV menjadi format yang dapat dibaca oleh PIL
img_pil = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

cv2_imshow(img)
```

# **NOMOR 2**
**lakukan edge detection menggunakan sobel dan canny**

```python
# Nomor 2
# Edge detection menggunakan Sobel
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)
cv2_imshow(sobel)
```

```python
# Edge detection menggunakan Canny
canny = cv2.Canny(img, 100, 200)
cv2_imshow(canny)
```

# **NOMOR 3**
**lakukan cropping: random cropping dan half-size center cropping**

```python
# Nomor 3
# Random Cropping
crop = transforms.RandomCrop(256)
crop(img_pil)
```

```python
# Half Center Cropping
center_x, center_y = img.shape[1] // 2, img.shape[0] // 2
cropped_center = img[center_y-50:center_y+50, center_x-50:center_x+50]
cv2_imshow(cropped_center)
```

## **NOMOR 4**
**lakukan transformasi pada gambar: putih jadi hitam, hitam jadi putih**

```python
# Nomor 4
# Hitam jadi Putih
image_black = cv2.bitwise_not(img)
cv2_imshow(image_black)
```

```python
# Putih Ke Hitam
image_white = cv2.bitwise_not(image_black)
cv2_imshow(image_white)
```

## ***NOMOR 5*
**tamahkan noise pada gambar original: salt and papper noise dan gaussian**

```python
# Nomor 5
# Salt and Pepper Noise
sp_noise = random_noise(gray_img, mode = 's&p', amount = 0.25)
sp_noise = sp_noise * 255
sp_noise = sp_noise.astype('uint8')
cv2_imshow(sp_noise)
```

```python
# Gaussian Noise
gaussian_noise = random_noise(gray_img, mode = 'gaussian', mean = 0, var = 0.075)
gaussian_noise = 255 * gaussian_noise
gaussian_noise = gaussian_noise.astype('uint8')
cv2_imshow(gaussian_noise)
```

## **NOMOR 6**
**lakukn transformasi geometri: rotasi, flip vertikal dan hosrisontal**

```python
# Nomor 6
# Rotasi
rotate = transforms.RandomRotation(15)
rotate(img_pil)
```

```python
# Flip Vertikal
vertikal = transforms.RandomVerticalFlip()
vertikal(img_pil)
```

```python
# Flip Horizontal
horizontal = transforms.RandomHorizontalFlip()
horizontal(img_pil)
```
